**DESCRIPCIÓN DEL CASO:** Se desea registrar en una lista los asistentes a un evento en un club, en un ambiente tipo CRUD. Los miembros del club tienen un número de socio, que es de 4 dígitos; cada asistente ha proporcionado el nombre como quiere que aparezca su diploma del evento; la aportación para ingresar al evento es voluntaria en su monto, pero obligatoria para cada asistente. La aportación mínima es de 100 pesos.

Al momento, se han registrado las sitguientes personas:

|Socio|Nombre|Aportación|
|---|---|---|
|1234|Juan|100.00|
|2345|María|400.00|
|3456|Brenda|350.00|

La llave en el diccionario será el número de socio (`socio`), y el valor del diccionario será una lista, que contenga el número de socio (`socio`), el nombre como aparecerá en el diploma (`nombre`), y la aportación realizada (`aportación`).

El programa debe permitir:
- Cargar en un diccionario inicial, llamado Asistentes, los socios que ya están inscritos.
- Seleccionar lo que se desea hacer, usando un menú.
- Agregar asistentes al diccionario.
    1. Registra los datos de un asistente nuevo. No validar, para no complicar el ejercicio. 
    1. No es posible registrar a dos asistentes con el mismo número de socio.
- Buscar asistentes en el diccionario.
    1. Buscar por número de socio.
    1. Si no encuentra lo buscado, se reporta que no existe.
    1. Si se encuentra lo buscado, muestra los datos correspondientes.
- Eliminar elementos en el diccionario.
    1. Buscar por número de socio.
    1. Si no existe el elemento, reporta que no es posible la eliminación.
    1. Si existe el elemento, lo elimina.
- Modificar asistentes en el diccionario.
    1. Buscar por número de socio.
    1. Si no encuentra lo buscado, se reporta que no existe.
    1. Si se encuentra lo buscado, muestra los datos correspondientes.
    1. Capura los datos que no son llave, y actualiza el elemento.
- Ver el contenido del diccionario.
    1. Muestra cuántos elementos hay en el diccionario.
    1. Muestra el contenido del diccionario.
    1. Muestra el total de aportaciones, sumarizado.
- Quitar todos los elementos de la lista.
- Salir del menú.

```python
# Declara un diccionario vacío, de nombre Asistentes
Asistentes={}

# Agrega los elementos existentes.
Asistentes.update({"1234":["1234","Juan",100.00]})
Asistentes.update({"2345":["2345","María",400.00]})
Asistentes.update({"3456":["3456","Brenda",350.00]})

# Muestra el contenido del diccionario.
print(Asistentes)
```

```python
# Menú general del programa.

print("\nMenú de opciones:")
print("\t[A] Agregar")
print("\t[B] Buscar")
print("\t[E] Eliminar elemento")
print("\t[M] Modificar elemento")
print("\t[V] Ver datos")
print("\t[Q] Quitar todos los elementos de la lista")
print("\t[S] Salir")

# Inicia un ciclo infinito, para controlar el programa.
while True:
    
    # Pregunta la opción de menú. Usa una variable
    # llamada opcion
    opcion=input("\n¿QUÉ DESEAS HACER?: ")
    
    # Valida si se capturó una opción válida (ABEMVQS)
    # o de lo contrario, muestra mensaje y vuelve a preguntar. 
    # Usa upper() para homologar a mayúsculas lo capturado, y usa
    # el operador in.
    if (not opcion.upper() in "ABEMVQS"):
        print("Opción no válida, intenta de nuevo.")
        continue

    # Opción: Salir
    # Si la opción es "S", imprime mensaje de salida, y se
    # sale del ciclo general usando break. 
    # Es el fin del programa.
    if (opcion.upper()=="S"):
        print("Fin de la ejecución.")
        break
            
    # Opción: Agregar asistente
    # Si la opción es "A", agrega un asistente.
    if (opcion.upper()=="A"):
        # Pregunta el dato llave.
        socio=input("Número de socio a agregar: ")
        # Si no hay un elemento con ese número de socio
        # se captura el resto de los datos, se agrega el
        # elemento, y se muestra mensaje.
        # De lo contrario, manda mensaje de existencia, y no
        # se puede agregar.
        # Usa get() para saber si hay coincidencias, y
        # update() para agregar elementos.
        if Asistentes.get(socio)==None:
            nombre=input("Nombre del socio: ")
            aportacion=float(input("Aportación: "))
            Asistentes.update({socio:[socio,nombre,aportacion]})
            print("Asistente agregado.")
        else:
            print("Asistente encontrado. No se puede repetir.")

    # Opción: Buscar asistente
    # Si la opción es "B", busca el asistente y muestra sus datos.
    if (opcion.upper()=="B"):
        # Pregunta el dato llave.
        socio=input("Número de socio a buscar: ")
        # Si no hay un elemento con ese número de socio, manda mensaje.
        # De lo contrario, muestra los datos.
        # Usa get() para vlidar coincidencias y mostrar datos.
        # Pon el resultado de get() en una variable llamada recuperado.
        recuperado=Asistentes.get(socio)
        if recuperado==None:
            print("Asistente no encontrado.")
        else:
            print(f"{recuperado[0]},{recuperado[1]},{recuperado[2]}")

    # Eliminar asistente
    # Si la opción es "E", busca el asistente y lo elimina.
    if (opcion.upper()=="E"):
        # Pregunta el dato llave.
        socio=input("Número de socio a eliminar: ")
        # Si el asistente no existe, mandar mensaje que no es posible eliminar.
        # De lo contrario, elimina el asistente y manda mensaje.
        # Usa get() para saber si hay coincidencias,
        # y pop(), para eliminar el elemento a partir de su llave.        
        if Asistentes.get(socio)==None:
            print("Asistente no encontrado. No se puede eliminar.")
        else:
            Asistentes.pop(socio)
            print("Asistente eliminado.")

    # Opción: Modificar asistente
    # Si la opción es "M", modifica los datos de un asistente
    # ya registrado.
    if (opcion.upper()=="M"):
        # Pregunta el dato llave.
        socio=input("Número de socio a modificar: ")
        # Si no hay un elemento con ese número de socio
        # manda mensaje de imposibilid de modificación.
        # Si existe, se muestran los datos existentes, se captura 
        # el resto de los datos, se actualiza el
        # elemento, y se muestra mensaje.
        # Usa get() para saber si hay coincidencias, usando una 
        # variable, y usa update() para actualizar el elemento.
        recuperado=Asistentes.get(socio)
        if recuperado==None:
            print("Asistente no encontrado. No se puede modificar")
        else:
            print(f"Datos actuales: {recuperado[0]},{recuperado[1]},{recuperado[2]}")
            nombre=input("Nuevo nombre del socio: ")
            aportacion=float(input("Nueva aportación: "))
            Asistentes.update({socio:[socio,nombre,aportacion]})
            print("Asistente modificado.")
        
    # Ver diccionario
    # Si la opción es "V", muestra cuántos elementos tiene el diccionario
    # y sus elementos. Se debe mostrar tabién la sumatoria de las aportaciones.
    if (opcion.upper()=="V"):
        # Declara ua variable para acumular las aportaciones.
        acumulado=0.0
        # Muestra cuántos elementos tiene la lista.
        # Usa len() para saber el número de elementos, y for,
        # para leer secuencialmente la lista. Usa el operador +=
        # para acumular.
        print(f"Elementos en la lista: {len(Asistentes)} ")
        for v in Asistentes.values():
            print(f"{v[0]},{v[1]},{v[2]}")
            acumulado+=v[2]
        print(f"Acumulado: ${acumulado:,.2f} ")

    # Quitar todos los elementos
    # Si la opción es "Q", elimina todos los elementos del diccionario
    # y manda mensaje.
    # Usa clear(), para eliminar todos los elementos del diccionario.
    if (opcion.upper()=="Q"):
        # Elimina todos los elementos de la lista
        Asistentes.clear()
        # Muestra mensaje
        print("Todos los asistentes eliminados.")
        
# Probar lo siguiente:
# Ver datos
# Agregar: "4567". "Diana", 800.00
# Buscar, "1234"
# Buscar, "5555"
# Eliminar, "2345"
# Eliminar, "5555"
# Ver lista
# Modificar "4567", con "Diana", y 600.00
# Ver lista
# Quitar todos los elementos de la lista
# Ver lista
# Salir
```
